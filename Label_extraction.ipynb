```python
from paddleocr import PaddleOCR, draw_ocr # main OCR dependencies
from matplotlib import pyplot as plt # plot images
import cv2 #opencv
import os # folder directory navigation
```

# 2. Instantiate Model and Detect

```python
# Setup model
ocr_model = PaddleOCR(lang='en',use_gpu=False)
# setting up which character model we actually want to use
```

```python
# setting the path of the img
#img_path = os.path.join('.', 'drug4.jpg')
img_path ="/drug9.jpg"
```

```python
# Run the ocr method on the ocr model
result = ocr_model.ocr(img_path)
```

# 3. Visualise Results

```python
# Extracting detected components
boxes = [res[0] for res in result[0]] #box coordinates
texts = [res[1][0] for res in result[0]]
scores = [res[1][1] for res in result[0]] #confidence metrix
```

```python
# Specifying font path for draw_ocr method
#font_path = os.path.join('PaddleOCR', 'doc', 'fonts', 'latin.ttf')
#font_path = "/content/PaddleOCR/doc/fonts/latin.ttf"
font_path = "/content/PaddleOCR/doc/fonts/french.ttf" #french
#path for eng detection (arabic)
```

```python
# Import our image - drug 1/2/3
img = cv2.imread(img_path) # imports image
#by default img in BGR-> RGB (OpenCV supports BGR model)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # reorders the color channels
```

```python
# Visualize our image and detections

# resizing display area
plt.figure(figsize=(15,15))

# draw annotations on image
annotated = draw_ocr(img, boxes, texts, scores, font_path=font_path) 

# show the image using matplotlib
plt.imshow(annotated) 
```
